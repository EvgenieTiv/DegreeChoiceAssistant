```python
import json

from market_retrieval import retrieve_market_context
```

```python
past_result = {
    "agent_name": "past_agent",
    "can_recommend": True,
    "degree_or_course": "course",
    "recommended_fields": [
        "computer_science"
    ],
    "not_recommended_fields": [],
    "uncertain_fields": [
        "chemistry",
        "mathematics",
        "physics"
    ],
    "confidence": "medium",
    "short_reason": "Strong technical and science ability but lacks high school graduation and prefers a new field."
}
```

```python
preferences_result = {
    "agent_name": "preferences_agent",
    "job_style_preference": "thinking",
    "work_environment_preference": "team",
    "preferred_field": "science",
    "self_learning_comfort": "yes",
    "long_learning_willingness": "yes",
    "learning_structure_preference": "structured",
    "openness_to_new_fields": "yes",
    "confidence": "high",
    "short_reason": "The user prefers analytical work, science-related directions, and a structured learning path."
}
```

```python
country = "Israel"
```

```python
retrieval_result = retrieve_market_context(
    country=country,
    past_analysis=past_result,
    preferences_analysis=preferences_result,
)
```

```python
print("Queries:")
for i, q in enumerate(retrieval_result["queries"], start=1):
    print(f"{i}. {q}")
```

```python
print(f"Total results: {len(retrieval_result['results'])}")
```
