```python
import json

from past_agent import build_past_profile, analyze_past_profile
```

```python
country = "Israel"
has_high_school_graduation = "no"
wants_to_continue_same_field = "no"
main_school_focus = "science"

advanced_subjects = [
    "computer_science",
    "physics",
    "mathematics",
    "chemistry",
]

favorite_subjects = [
    "computer_science",
    "physics",
]

best_subjects = [
    "computer_science",
    "mathematics",
]
```

```python
past_profile = build_past_profile(
    country=country,
    has_high_school_graduation=has_high_school_graduation,
    wants_to_continue_same_field=wants_to_continue_same_field,
    main_school_focus=main_school_focus,
    advanced_subjects=advanced_subjects,
    favorite_subjects=favorite_subjects,
    best_subjects=best_subjects,
)
```

```python
print(json.dumps(past_profile, ensure_ascii=False, indent=2))
```

```python
past_result = analyze_past_profile(past_profile)
```

```python
print(json.dumps(past_result, ensure_ascii=False, indent=2))
```
