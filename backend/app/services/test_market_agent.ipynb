```python
import json

from market_agent import build_market_profile, analyze_market_profile
```

```python
past_result = {
    "agent_name": "past_agent",
    "can_recommend": True,
    "degree_or_course": "course",
    "recommended_fields": [
        "computer_science"
    ],
    "not_recommended_fields": [],
    "uncertain_fields": [
        "chemistry",
        "mathematics",
        "physics"
    ],
    "confidence": "medium",
    "short_reason": "Strong technical and science ability but lacks high school graduation and prefers a new field."
}
```

```python
preferences_result = {
    "agent_name": "preferences_agent",
    "job_style_preference": "thinking",
    "work_environment_preference": "team",
    "preferred_field": "science",
    "self_learning_comfort": "yes",
    "long_learning_willingness": "yes",
    "learning_structure_preference": "structured",
    "openness_to_new_fields": "yes",
    "confidence": "high",
    "short_reason": "The user prefers analytical work, science-related directions, and a structured learning path."
}
```

```python
country = "Israel"
```

```python
market_context = """
Israel labor market summary:
- Entry-level tech support, QA, and data-related roles show recurring demand.
- Data analysis and business operations roles are often more accessible than pure research tracks.
- Pure science roles such as chemistry and physics often have narrower entry paths and more formal education requirements.
- Adjacent transitions from technical backgrounds into operations or management-related paths can be realistic.
- Practical market value is stronger for roles with shorter training paths than for highly academic specializations.
"""
```

```python
market_profile = build_market_profile(
    country=country,
    past_analysis=past_result,
    preferences_analysis=preferences_result,
    market_context=market_context,
)
```

```python
print(json.dumps(market_profile, ensure_ascii=False, indent=2))
```

```python
market_result = analyze_market_profile(market_profile)
```

```python
print(json.dumps(market_result, ensure_ascii=False, indent=2))
```

```python
print("Agent:", market_result["agent_name"])
print("Can recommend:", market_result["can_recommend"])
print("Confidence:", market_result["confidence"])
print("Reason:", market_result["short_reason"])
```

```python
print("Recommended paths:")
for i, item in enumerate(market_result["recommended_paths"], start=1):
    print(f"{i}. Field:", item["field"])
    print("   Career outcomes:", ", ".join(item["career_outcomes"]))
    print("   Reason:", item["reason"])
```

```python
print("Uncertain paths:")
for i, item in enumerate(market_result["uncertain_paths"], start=1):
    print(f"{i}. Field:", item["field"])
    print("   Career outcomes:", ", ".join(item["career_outcomes"]))
    print("   Reason:", item["reason"])
```
