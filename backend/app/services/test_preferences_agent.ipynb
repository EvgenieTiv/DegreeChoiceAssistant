```python
import json

from preferences_agent import build_preferences_profile, analyze_preferences_profile
```

```python
job_style_preference = "thinking"
work_environment_preference = "team"
preferred_field = "science"
self_learning_comfort = "yes"
long_learning_willingness = "yes"
learning_structure_preference = "structured"
openness_to_new_fields = "yes"
```

```python
preferences_profile = build_preferences_profile(
    job_style_preference=job_style_preference,
    work_environment_preference=work_environment_preference,
    preferred_field=preferred_field,
    self_learning_comfort=self_learning_comfort,
    long_learning_willingness=long_learning_willingness,
    learning_structure_preference=learning_structure_preference,
    openness_to_new_fields=openness_to_new_fields,
)
```

```python
print(json.dumps(preferences_profile, ensure_ascii=False, indent=2))
```

```python
preferences_result = analyze_preferences_profile(preferences_profile)
```

```python
print(json.dumps(preferences_result, ensure_ascii=False, indent=2))
```

```python
print("Agent:", preferences_result["agent_name"])
print("Job style preference:", preferences_result["job_style_preference"])
print("Work environment preference:", preferences_result["work_environment_preference"])
print("Preferred field:", preferences_result["preferred_field"])
print("Self learning comfort:", preferences_result["self_learning_comfort"])
print("Long learning willingness:", preferences_result["long_learning_willingness"])
print("Learning structure preference:", preferences_result["learning_structure_preference"])
print("Openness to new fields:", preferences_result["openness_to_new_fields"])
print("Confidence:", preferences_result["confidence"])
print("Reason:", preferences_result["short_reason"])
```
